```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)


# Combine the data into a single dataset

merge_data = city_data.merge(ride_data, how='outer', on='city')


# Display the data table for preview
merge_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

type_group = merge_data.groupby(['type', 'city']).mean()
type_group['number_of_rides'] = merge_data.groupby(['type', 'city'])['ride_id'].count()
type_group['total_fare'] = merge_data.groupby(['type', 'city'])['fare'].sum()
type_group.head()

```

```python
# Build the scatter plots for each city types
urban_x = type_group.loc['Urban']['number_of_rides'].values
suburban_x = type_group.loc['Suburban']['number_of_rides'].values
rural_x = type_group.loc['Rural']['number_of_rides'].values

urban_y = type_group.loc['Urban']['fare'].values
suburban_y = type_group.loc['Suburban']['fare'].values
rural_y = type_group.loc['Rural']['fare'].values

urban_s = type_group.loc['Urban']['driver_count'].values * 10
suburban_s = type_group.loc['Suburban']['driver_count'].values * 10
rural_s = type_group.loc['Rural']['driver_count'].values * 10

plt.scatter(urban_x, urban_y, label='Urban', s=urban_s, alpha=.8, c='LightCoral', edgecolors='black', linewidths=1)
plt.scatter(suburban_x, suburban_y, label='Suburban', s=suburban_s, alpha=.8, c='LightSkyBlue', edgecolors='black', linewidths=1)
plt.scatter(rural_x, rural_y, label='Rural', s=rural_s, alpha=.8, c='Gold', edgecolors='black', linewidths=1)


# Incorporate the other graph properties

plt.title('Pyber Ride Sharing Data (2016)')
plt.ylabel('Average Fare ($)')
plt.xlabel('Total Number of Rides (Per City)')
plt.grid()

# Create a legend

lgnd = plt.legend(title='City Types')
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size

plt.text(42, 35, 'Note:\nCircle size correlates with driver count per city')

# Save Figure

plt.savefig('Pyber_Scatter.png', bbox_inches='tight')
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_fares = type_group.loc['Urban']['total_fare'].sum()
rural_fares = type_group.loc['Rural']['total_fare'].sum()
suburban_fares = type_group.loc['Suburban']['total_fare'].sum()

labels = ['Urban', 'Rural','Suburban']
explode = [0.1, 0, 0]
colors = ['LightCoral', 'Gold', 'LightSkyBlue']

# Build Pie Chart

plt.pie([urban_fares, rural_fares, suburban_fares], explode=explode, colors=colors, labels=labels, autopct="%1.1f%%", shadow=True, startangle=300)
plt.axis("equal")
plt.title('% of Total Fares by City Type')

# Save Figuresuburban_fares
plt.savefig('Fares_Pie.png', bbox_inches='tight')
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_rides = type_group.loc['Urban']['number_of_rides'].sum()
rural_rides = type_group.loc['Rural']['number_of_rides'].sum()
suburban_rides = type_group.loc['Suburban']['number_of_rides'].sum()

labels = ['Urban', 'Rural','Suburban']
explode = [0.1, 0, 0]
colors = ['LightCoral', 'Gold', 'LightSkyBlue']

# Build Pie Chart
plt.pie([urban_rides, rural_rides, suburban_rides], explode=explode, colors=colors, labels=labels, autopct="%1.1f%%", shadow=True, startangle=260)
plt.axis("equal")
plt.title('% of Total Rides by City Type')

# Save Figure
plt.savefig('Rides_Pie.png', bbox_inches='tight')
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
urban_drivers = type_group.loc['Urban']['driver_count'].sum()
rural_drivers = type_group.loc['Rural']['driver_count'].sum()
suburban_drivers = type_group.loc['Suburban']['driver_count'].sum()

labels = ['Urban', 'Rural','Suburban']
explode = [0.1, 0, 0]
colors = ['LightCoral', 'Gold', 'LightSkyBlue']

# Build Pie Charts
plt.pie([urban_drivers, rural_drivers, suburban_drivers], explode=explode, colors=colors, labels=labels, autopct="%1.1f%%", shadow=True, startangle=230)
plt.axis("equal")
plt.title('% of Total Drivers by City Type')

# Save Figure
plt.savefig('Drivers_Pie.png', bbox_inches='tight')
```

```python
# Show Figure
plt.show()
```
